## 递归、递推

### 递推

````python
def sum_list_iter(x):
    sum = 0
    for num in x:
        sum += num
        print(sum, end=" ")
    return sum

sum_list_iter([1, 2, 3, 4, 5])
````

### 递归

````python
def sum_list_recur(x):
    if not x:
        return 0
    else:
        s = x[0] + sum_list_recur(x[1:])
        print(s, end=" ")
        return s
    
sum_list_recur([1, 2, 3, 4, 5])
````

````python
def sum_list(x, sum=0):
    if not x:
        return sum
    else:
        print(sum + x[0], end=" ")
        return sum_list(x[1:], sum + x[0])
    
sum_list([1, 2, 3, 4, 5])
````

## 栈、树与递归

````python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

factorial(5)
````

````python
def fibonacci(n):
	if n == 0:
		return 0
	elif n == 1:
		return 1
	return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(6) # 1 1 2 3 5 8
````

````python
def fibonacci_iterative(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    prev1, prev2 = 1, 0
    for _ in range(2, n + 1):
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current
    
    return prev1

print(fibonacci_iterative(6))
````
